```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_url = "https://people.stfx.ca/dmattie/bsad482/iris.csv"
iris_df = pd.read_csv(iris_url)
```

```python
plt.figure(figsize=(10, 8))
sns.pairplot(iris_df, hue="Species", diag_kind="hist")
plt.suptitle("Pair Plot of the Iris Dataset", y=1.02)
plt.show()
```

```python
mtcars_url = "https://people.stfx.ca/dmattie/bsad482/mtcars.csv"
mtcars_df = pd.read_csv(mtcars_url)
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="cyl", y="mpg", data=mtcars_df)
plt.title("Violin Plot of MPG by Number of Cylinders")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
weather_url = "https://people.stfx.ca/dmattie/bsad482/weather.csv"
weather_df = pd.read_csv(weather_url)
```

```python
weather_df['date'] = pd.to_datetime(weather_df['date'])
```

```python
weather_df['month'] = weather_df['date'].dt.strftime('%B')
```

```python
months_order = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']
```

```python
plt.figure(figsize=(10, 8))
sns.set(style="white")
```

```python
g = sns.FacetGrid(weather_df, row="month", hue="month", aspect=4, height=1.2, palette="coolwarm", row_order=months_order)
g.map(sns.kdeplot, "tmax", fill=True, alpha=0.7)
g.set_titles("{row_name}")
g.set_xlabels("Max Temperature (°C)")
g.set_yticklabels([])
plt.subplots_adjust(hspace=0.5)
plt.show()
```
